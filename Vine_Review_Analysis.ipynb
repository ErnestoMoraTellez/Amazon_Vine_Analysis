```python
 # Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("M16-Amazon-Challenge-2").getOrCreate()
```

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Sports_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
# Create the vine_table. DataFrame
vine_df = df.select(['review_id','star_rating','helpful_votes','total_votes','vine','verified_purchase'])
```

```python
vine_df.show()
```

```python
total_votes_df = vine_df.filter('total_votes>=20')
total_votes_df.show()
```

```python
helpful_votes_df = total_votes_df.filter('(helpful_votes/total_votes)>=0.5')
helpful_votes_df.show()
```

```python
y_vine_helpful_votes_df = helpful_votes_df.filter('vine="Y"')
y_vine_helpful_votes_df.show()
```

```python
n_vine_helpful_votes_df = helpful_votes_df.filter('vine="N"')
n_vine_helpful_votes_df.show()
```

```python
y_vine_total_reviews = y_vine_helpful_votes_df.count()
print(y_vine_total_reviews)
```

```python
y_vine_total_reviews_5_stars = y_vine_helpful_votes_df.filter('star_rating=5').count()
print(y_vine_total_reviews_5_stars)
```

```python
y_percentage_paid_reviews = y_vine_total_reviews_5_stars/y_vine_total_reviews
print(y_percentage_paid_reviews)
```

```python
n_vine_total_reviews = n_vine_helpful_votes_df.count()
print(n_vine_total_reviews)
```

```python
n_vine_total_reviews_5_stars = n_vine_helpful_votes_df.filter('star_rating=5').count()
print(n_vine_total_reviews_5_stars)
```
